## Teller Functions ("tell_")

Teller functions are functions which return **the most interesting insights about the COVID-19 pandemic, many of which cannot be instantly worked out from the raw databases returned by Getter functions.** For example, get_vaccine_admin returns a DataFrame with plenty of data about COVID-19 vaccinations, but among them there is not the *total amount of people who have completed the vaccination cycle*: even though this is of great importance, the user would be forced to manually calculate this value with proper operations on the DataFrame, also taking into account requirements such as the fact that Janssen vaccine only requires one dose, as does every person who has been infected by SARS-CoV-2 before getting the vaccine. Teller functions do this for you, returning the end result (referring to the example, this is done by tell_total_vaccinated) and integrating Getter functions for those important values which, without the former, would be unjustifiably hard to get.


### tell_manufacturer_delivered_doses(manufacturer)

Returns the number of delivered vaccine doses from the manufacturer given as a parameter in Italy as an integer. If string "all" is provided as parameter, it returns the number of delivered vaccine doses from all manufacturers.
ONLY ACCEPTED MANUFACTURERS AND SPELLINGS ARE "Pfizer/BioNTech", "Moderna", "Vaxzevria (AstraZeneca)" AND "Janssen". Strings other than manufacturer names or "all" yield an error.

```python
icl.tell_manufacturer_delivered_doses("Pfizer/BioNTech")
```

### tell_total_vaccinated(dose_number, option = "n")

Depending on the integer value provided as *dose_number*:
- dose_number = 1: returns the number of individuals who have been injected at least one vaccine dose in Italy (independently of it being enough for vaccination cycle completion, as is the case with Janssen vaccine or for individuals with recent COVID-19 injection, for whom only one dose is required);
- dose_number = 2: returns the number of individuals who have completed the vaccination cycle in Italy (with double dose for Pfizer/BioNTech, Moderna and Vaxzevria (AstraZeneca), with single dose for Janssen, with single dose for individuals previously infected with COVID-19 between 3 and 6 months before vaccination);
- dose_number = 3: returns the number of individuals who have been injected an extra dose of vaccine in Italy, being eligible for it depending on their medical condition (commonly referred to as "third dose" in media).

Result is returned depending on the string value provided as *option*:
- no option or option = "n" or "number": just returns the requested number as an integer (default option);
- option = "o" or "over12": returns the requested number as a fraction of the Italian population aged over 12 (to which many official documents refer to) as a floating point;
- option = "p" or "population": returns the requested number as a fraction of the whole Italian population as a floating point.

**WARNING: function might take a while before returning!**

The following example returns the total number of people who have completed the vaccination cycle:

```python
icl.tell_total_vaccinated(2)
```

This other example returns the percentage of individuals aged over 12 who have completed the vaccination cycle. This is a value usually quoted in media (since the official purpose of the vaccination campaign, as stated by Gen. Figliuolo, Extraordinary Commissioner for the COVID-19 emergency, is reaching 80% of over 12 individuals fully vaccinated by 30th September).

```python
# returns a fraction, not a percentage
fraction_of_total_vaccinated = icl.tell_total_vaccinated(dose_number=2, option="over12")
percentage_of_total_vaccinated = fraction_of_total_vaccinated*100
print("{}%".format(percentage_of_total_vaccinated))
```

#### Full documentation

```python
help(icl.tell_total_vaccinated)
```

## Grapher Functions

Common Python libraries like Matplotlib provide an easy way for plotting data. However, this may become difficult when one wants to plot data on a map, since they would also need a complete description of political entities and borders. **Grapher functions make plotting on a map as easy as plotting a histogram.**
NOTE: they are only guaranteed to work with Italian COVID Library generated DataFrames. Moreover, said DataFrames must have data referred to regions or provinces (depending if one aims at plotting on a map of regions or provinces).

### plot_on_map(source, on, column, legend = True, cmap = "viridis")

Plots data coming from column *column* of Italian COVID Library-generated DataFrame *source* on a map of Italy with regions or provinces, depending on the option *on* specified (options "region", "regions" or "r" for plotting on regions, options "province", "provinces" or "p" for provinces), displaying or not a legend (depending on option *legend*, default is *True*) and with the color palette *cmap* (color palettes are matplotlib ones, default is *viridis*).

NOTE 1: DataFrame must contain regional or provincial data, only "getter" functions referred to regional or provincial phenomena return plotting-compatible DataFrames.

NOTE 2: if you want to use the full customization power of libraries like GeoPandas, head to prepare_for_plotting_on_map.

The following example plots the number of new COVID-19 cases per region on a map of Italian regions, displaying a legend and using the color palette *Reds*:

```python
# gets the DataFrame with the data of our interest
cases_per_region = icl.get_region_cases(latest=True)
# new cases data is in the column "new_cases" of our DataFrame
cases_per_region_plot = icl.plot_on_map(source=cases_per_region, on="regions", column="new_cases", legend=True, cmap="Reds")
```

#### Full documentation

```python
help(icl.plot_on_map)
```

### prepare_for_plotting_on_map(source, on)

Makes Italian COVID Library-generated DataFrame *source* compatible with GeoPandas (i.e. a GeoDataFrame), for subsequent plotting on a map with Italian regions or provinces, depending on the option *on* specified (options "region", "regions" or "r" for plotting on regions, options "province", "provinces" or "p" for provinces). Use this function if you want to display your data on a map using the full potential of GeoPandas.

NOTE: DataFrame must contain regional or provincial data, only "getter" functions referred to regional or provincial phenomena return potential GeoPandas-compatible DataFrames.

The following example makes the DataFrame generated by get_region_cases_latest, containing the number of new COVID-19 cases per region, compatible with GeoPandas in order to subsequently plot these data on a map of Italian regions:

```python
# gets the DataFrame with the data of our interest
cases_per_region = icl.get_region_cases(latest=True)
# prepare_for_plotting_on_map turns a Pandas DataFrame into a GeoPandas GeoDataFrame
cases_per_region_geopandas_compatible = icl.prepare_for_plotting_on_map(source=cases_per_region, on="regions")
```

We can now import GeoPandas and use it to plot data on a map of Italy with its regions. The following example in particular outputs the very same result as the example shown for the function plot_on_map: if you only need simple customization, head to plot_on_map which, given an Italian COVID Library-generated DataFrame, directly plots it on a map; prepare_for_plotting_on_map is better if you aim at more advanced customization using GeoPandas.

```python
import geopandas
# below the function plot of GeoPandas is used
# new cases data is in the column "new_cases" of both the original DataFrame and its GeoPandas compatible version
cases_per_region_plot = cases_per_region_geopandas_compatible.plot(column="new_cases", legend=True, cmap="Reds")
```

#### Full documentation

```python
help(icl.prepare_for_plotting_on_map)
```

## Use examples of Italian COVID Library

The following shows some practical use examples of Italian COVID Library functions.

### Plotting the trend of COVID-19 infections since the beginning of the pandemic

In this example we shall see how to display a plot with the trend of COVID-19 infections (i.e. the number of new cases each day) from the first cases in February 2020.

```python
import pandas as pd
import matplotlib.pyplot as plt
# new cases per day are returned by get_national_trend
new_cases_trend = icl.get_national_trend()
plt.plot(new_cases_trend.index, new_cases_trend["new_cases"])
```

The output is a graph with on the x axis the date and on the y axis the number of new cases diagnosed on that day. You can notice the four waves of the epidemic in correspondence of the peaks: the first one of Spring 2020, the second one in Autumn 2020, the third Alpha variant-induced one in Spring 2021 and the fourth Delta variant-induced one in late Summer 2021. Cases numbers are low during the first wave because of the poor and immature detection techniques used back then. You can also notice the high frequency fluctuation in data: this is because during weekends cases are reported slowlier, ending up being registered the following days and reducing the amount of cases officially reported for weekends.

### Getting and presenting the vaccination situation of individuals aged 50-59

In this example we shall see how to evaluate the vaccination situation of individuals in their fifties. One of current (September-October 2021) issues of the pandemic is, in fact, the still high percentage of people aged between 50 and 59 who have not received any dose of the vaccine, posing a further obstacle on the so-yearned for return to normalcy. We also want to present the situation as a pie chart.

```python
import matplotlib.pyplot as plt

# we need vaccine administration data
vaccine_admin = icl.get_vaccine_admin()
# we are only interested in data concerning people aged 50-59
vaccine_admin_50_59 = vaccine_admin[vaccine_admin["age_group"]=="50-59"]
# Janssen only requires one dose, so do previously infected individuals. Individuals given the 3rd dose are counted separately
fully_vaccinated_50_59 = vaccine_admin_50_59[vaccine_admin_50_59["manufacturer"]!="Janssen"].sum()["second_dose"]+vaccine_admin_50_59[vaccine_admin_50_59["manufacturer"]=="Janssen"].sum()["first_dose"]+vaccine_admin_50_59.sum()["previously_infected"]-vaccine_admin_50_59.sum()["extra_dose"]
extra_dose_vaccinated_50_59 = vaccine_admin_50_59.sum()["extra_dose"]
singly_vaccinated_50_59 = vaccine_admin_50_59.sum()["first_dose"]+vaccine_admin_50_59.sum()["previously_infected"]-fully_vaccinated_50_59-extra_dose_vaccinated_50_59
# we also need the amount of people to be vaccinated
eligible_data = icl.get_eligible()
# again, we are only interested in data concerning people aged 50-59
eligible_data_50_59 = eligible_data[eligible_data.index=="50-59"]
# sum is performed because data are returned separated per region
eligible_50_59 = eligible_data_50_59.sum()["population"]
non_vaccinated_50_59 = eligible_50_59-fully_vaccinated_50_59-extra_dose_vaccinated_50_59-singly_vaccinated_50_59
# we now have all the data we wanted
print("Fully vaccinated (excluding individuals given extra dose): {}".format(fully_vaccinated_50_59))
print("People also given extra dose: {}".format(extra_dose_vaccinated_50_59))
print("Singly vaccinated: {}".format(singly_vaccinated_50_59))
print("Non vaccinated: {}".format(non_vaccinated_50_59))
print("TOTAL: {}".format(eligible_50_59))

# we can now plot these data. Best way is through a pie chart
labels = "Fully, also given 3rd dose", "Fully", "Partially", "No injection"
# "sizes" requires percentages, calculated over amount of individuals eligible for vaccination
sizes = [extra_dose_vaccinated_50_59/eligible_50_59*100, fully_vaccinated_50_59/eligible_50_59*100, singly_vaccinated_50_59/eligible_50_59*100, non_vaccinated_50_59/eligible_50_59*100]
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```

We see (October 2021) that the vast majority of people aged 50-59 are vaccinated, but there is still a relevant group still to get their first dose. People in their fifties are indeed considered in danger of severe COVID, so that is why there being even such a relatively small percentage of non vaccinated individuals is risky.

### Plotting the variation of Rt index over time

The R index is defined as the average number of individuals a person positive to COVID-19 can infect. When health measures are in place such as social distancing, this value is called the Rt index. In this example we shall see how to get it and plot it over time.

First, let us get the Rt values.

```python
rt_values = icl.tell_rt()
rt_values
```

We are now ready to plot these values.

Acknowledgements: this is an adaptation of a use example suggested for epyestim, the library tell_rt uses to calculate Rt values.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,1, figsize=(12, 4))

rt_values.loc[:,"Q0.5"].plot(ax=ax, color="red")
ax.fill_between(rt_values.index, 
                    rt_values["Q0.025"], 
                    rt_values["Q0.975"], 
                    color="red", alpha=0.2)
ax.set_xlabel("Date")
ax.set_ylabel("R(t) with 95% confidence interval")
ax.set_ylim([0,3])
ax.axhline(y=1)
ax.set_title("Estimate of time-varying effective reproduction number for Italy")
plt.show()
```

You can notice the peaks corresponding to the four waves of the pandemic, all of which with Rt above 1. The light red area is the error on the values with 95% confidence interval.

**WARNING: these Rt values might slightly differ from the ones given by other institutions or media outlets, depending on the algorithm they specifically use.**
